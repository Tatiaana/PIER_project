```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('/home/localadmin/Downloads/Minerals_Database.csv')
df
```

```python
df.Uranium.value_counts()
```

```python
df['Radioactive'] = df['Uranium'] + df['Thorium'] + df['Radium'] + df['Polonium'] + df['Plutonium'] + df['Americium']+df['Curium']+df['Astatine']+df['Neptunium']+df['Francium']+df['Livermorium']+df['Flerovium']+df['Actinium']+df['Moscovium']+df['Einsteinium']+df['Fermium']+df['Promethium']+df['Nihonium']+df['Californium']+df['Protactinium']+df['Bohrium']+df['Technetium']+df['Berkelium']+df['Radon']
```

```python
rows_with_rad = np.where(df[df != '0.0'].Radioactive)[0]
df = df.drop(rows_with_rad)
df = df.drop(columns=['Radioactive','Uranium','Thorium','Radium','Polonium','Plutonium','Americium','Curium','Astatine','Neptunium','Francium','Livermorium','Flerovium','Actinium','Moscovium','Einsteinium','Fermium','Promethium','Nihonium','Californium','Protactinium','Bohrium','Technetium','Berkelium','Radon'])
```

```python
df.index = df.Name
df = df.drop(columns=['Unnamed: 0','Name'])
df = df.drop(columns=np.concatenate([df.columns[7:116]]))
df
```

```python
df = df[(df.Dispersion>0.0)&(df['Crystal Structure']>0.0)&(df.Diaphaneity>0.0)&(df.Optical>0.0)]
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
sns.barplot(x='Crystal Structure', y='Mohs Hardness', data=df)
plt.xticks(rotation=90)
plt.xlabel('Crystal Structure')
plt.ylabel('Mohs Hardness')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Crystal Structure', y='Refractive Index', data=df)
plt.xticks(rotation=90)
plt.xlabel('Crystal Structure')
plt.ylabel('Refractive Index')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Crystal Structure', y='Optical', data=df)
plt.xticks(rotation=90)
plt.xlabel('Crystal Structure')
plt.ylabel('Optical')
plt.show()
```
