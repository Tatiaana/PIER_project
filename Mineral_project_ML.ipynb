```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('/home/localadmin/Downloads/Minerals_Database.csv')
df
```

```python
df = df.drop(columns=np.concatenate([df.columns[9:140]]))
df
```

```python
dataset = df.drop(columns=['Unnamed: 0','Name','Mohs Hardness','Diaphaneity','Specific Gravity','Refractive Index','Dispersion'])
dataset
```

```python
dataset1 = dataset[(dataset.Optical>0.0)&((dataset['Crystal Structure']>0.0))]
dataset1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
sns.barplot(x='Crystal Structure', y='Optical', data=dataset1)
plt.xticks(rotation=90)
plt.xlabel('Crystal Structure')
plt.ylabel('Optical')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
data = dataset1.drop('Optical', axis=1)
labels = dataset1.Optical
X_train, X_test, y_train, y_test = train_test_split(data, 
                                                    labels, 
                                                    random_state=0, 
                                                    test_size=0.2) 
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(100, ), 
                    max_iter=480, 
                    alpha=1e-4,
                    solver='sgd', 
                    tol=1e-4, 
                    random_state=1,
                    learning_rate_init=.1)


mlp.fit(X_train, y_train)
print("Training set score: %f" % mlp.score(X_train, y_train))
print("Test set score: %f" % mlp.score(X_test, y_test))
```
